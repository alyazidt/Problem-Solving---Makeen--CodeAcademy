```python
"""
Problem of the day #2
Given two integers white and black representing the count of white and black balls. return the maximum height of the triangle that can be achieved.
        
        (each triangle should start with white)

                (Note: each row should have the same color)

                        (Constraint --->  1 <= black , white <= 100)


                                Example #1 : 
                                                input :                                        
                                                        white  = 4
                                                        black  = 2
                                                
                                                output :
                                                        output = 3


                                                              ⚪
                                                            ⚫  ⚫
                                                          ⚪  ⚪  ⚪

----------------------------------------------------------------------------------------------------------------------------------------------------------

                                Example #2 : 
                                                input :
                                                        white  = 1
                                                        black  = 2
                                                
                                                output :
                                                        output = 2


                                                              ⚪
                                                            ⚫  ⚫

----------------------------------------------------------------------------------------------------------------------------------------------------------

                                Example #3 : 
                                                input :
                                                        white  = 1
                                                        black  = 1
                                                
                                                output :
                                                        output = 1    


                                                              ⚪

----------------------------------------------------------------------------------------------------------------------------------------------------------

                                Example #4 : 
                                                input :
                                                        white  = 1
                                                        black  = 10
                                                
                                                output :
                                                        output = 2        


                                                              ⚪
                                                            ⚫  ⚫

"""
```

```python

#Solve No.1 (Using Python)


white = 4
black = 2
height = 0

for i in range(1, 100):
    if i % 2 == 0:  #black
        if black >= i:
            black -= 1
            height += 1

            print(' ' * (2 * (7 - i)) + '  '.join(['⚫'] * i))

        else:
            break


    else:  #white
        if white >= i:
            white -= 1
            height += 1

            print(' ' * (2 * (7 - i)) + '  '.join(['⚪'] * i))
            
        else:
            break

print("\n The height is", height)
```

```python

#Solve No.2 (Using Python)


white = int(input("Enter White balls: "))
black = int(input("Enter black balls: "))
height = 0

print("The white ball : ", white)
print("The black ball : ", black ,"\n")

for i in range(1, 100):
    if i % 2 == 0:  #black
        if black >= i:
            black -= 1
            height += 1

            print(' ' * (2 * (7 - i)) + '  '.join(['⚫'] * i))

        else:
            break


    else:  #white
        if white >= i:
            white -= 1
            height += 1

            print(' ' * (2 * (7 - i)) + '  '.join(['⚪'] * i))
            
        else:
            break


print("\nThe white ball left : ", white)
print("The black ball left : ", black)
print("The height is", height)
```
